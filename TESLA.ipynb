```python
# Imports & Configuration
import os
import re
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from sklearn.decomposition import PCA

from pinecone import Pinecone, ServerlessSpec
from langchain_pinecone import PineconeVectorStore
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.document_loaders import PyPDFLoader

# --- CONFIGURATION ---
PINECONE_API_KEY = "api_key"
PINECONE_INDEX_NAME = "ms"

pc = Pinecone(api_key=PINECONE_API_KEY)
print("Pinecone connected")
```

```python
from google.colab import files
import pdfplumber
from langchain_core.documents import Document

# Re-upload the file
uploaded = files.upload()
pdf_filename = list(uploaded.keys())[0]

# Check it uploaded properly
import os
size_mb = os.path.getsize(pdf_filename) / (1024 * 1024)
print(f"File size: {size_mb:.2f} MB")

# Load with pdfplumber as before
docs = []
with pdfplumber.open(pdf_filename) as pdf:
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        if text and len(text.strip()) > 50:
            docs.append(Document(
                page_content=text,
                metadata={"page": i + 1, "source": pdf_filename}
            ))

print(f"Loaded {len(docs)} pages")
print(f"\nSample:\n{docs[0].page_content[:300]}")
```

```python
# Chunking
splitter = RecursiveCharacterTextSplitter(
    chunk_size=2000,
    chunk_overlap=400,
    separators=["\n\n", "\n", ". ", " ", ""]
)
splits = splitter.split_documents(docs)

# Enrich metadata
for i, chunk in enumerate(splits):
    chunk.metadata['chunk_id'] = i
    chunk.metadata['word_count'] = len(chunk.page_content.split())

# Remove near-empty chunks (headers, footers)
splits = [c for c in splits if c.metadata['word_count'] >= 30]

print(f"{len(splits)} chunks created")
print(f"   Avg words: {np.mean([c.metadata['word_count'] for c in splits]):.0f}")
```

```python
# Embedd and store in pinecone
# Load embedding model (free, runs locally)
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs={'device': 'cpu'},
    encode_kwargs={'normalize_embeddings': True}
)

# Create Pinecone index if it doesn't exist
existing = [idx.name for idx in pc.list_indexes()]
if PINECONE_INDEX_NAME not in existing:
    pc.create_index(
        name=PINECONE_INDEX_NAME,
        dimension=384,
        metric="cosine",
        spec=ServerlessSpec(cloud="aws", region="us-east-1")
    )
    print(f"Created index: {PINECONE_INDEX_NAME}")

index = pc.Index(PINECONE_INDEX_NAME)
vectorstore = PineconeVectorStore(index=index, embedding=embeddings)

# Upload in batches
for i in range(0, len(splits), 50):
    vectorstore.add_documents(splits[i:i+50])
    print(f"   Batch {i//50 + 1} uploaded")

print(f"\n{len(splits)} chunks stored in Pinecone!")
```

```python
ESG_QUERIES = {
    "Climate Risk":       "physical climate risk floods droughts extreme weather",
    "Carbon Emissions":   "greenhouse gas emissions Scope 1 Scope 2 CO2",
    "Supply Chain":       "supply chain supplier sourcing raw materials",
    "Water Usage":        "water consumption withdrawal usage intensity",
    "Governance":         "board diversity ESG governance oversight",
    "Net Zero / Targets": "net zero carbon neutral 2030 2050 targets",
    "Energy":             "renewable energy solar wind electricity consumption",
    "Labor & Human Rights": "human rights labor standards working conditions",
}

print(" Searching for ESG topics...\n")
topic_results = {}
for topic, query in ESG_QUERIES.items():
    results = vectorstore.similarity_search(query, k=3)
    topic_results[topic] = results
    pages = [str(r.metadata.get('page', '?')) for r in results]
    print(f"  --> {topic}: {len(results)} chunks (pages: {', '.join(pages)})")
```

```python
# Summarize
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

print("Loading Flan-T5-large...")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large")
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)
print(f"Model on {device}")
```

```python
import textwrap
def summarize(text, topic="ESG"):
    if len(text.split()) < 30:
        return "[Too short to summarize]"
    prompt = (
        f"You are an ESG analyst. Summarize the key {topic} data, "
        f"metrics, and commitments from this disclosure: {text[:1500]}"
    )
    inputs = tokenizer(prompt, return_tensors="pt", max_length=1024, truncation=True).to(device)
    outputs = model.generate(**inputs, max_length=150, min_length=40, num_beams=4, no_repeat_ngram_size=3)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Summarize each topic
summaries = {}
print("\n" + "\n ESG SUMMARIES\n" + "-"*50)
for topic, docs in topic_results.items():
    text = " ".join([d.page_content for d in docs[:2]])
    summary = summarize(text, topic)
    summaries[topic] = summary
    print(f"\n --> {topic.upper()}\n    {textwrap.fill(summary, width=80)}")
```

```python
# 3D visual of topic clusters
vectors, labels, pages, previews = [], [], [], []

for topic, docs in topic_results.items():
    for doc in docs:
        vec = embeddings.embed_query(doc.page_content[:300])
        vectors.append(vec)
        labels.append(topic)
        pages.append(str(doc.metadata.get('page', '?')))
        previews.append(doc.page_content[:100].replace('\n', ' '))

vecs_3d = PCA(n_components=3).fit_transform(np.array(vectors))

fig = px.scatter_3d(
    x=vecs_3d[:, 0], y=vecs_3d[:, 1], z=vecs_3d[:, 2],
    color=labels,
    hover_data={"Page": pages, "Preview": previews},
    title="ESG Topic Clusters in Semantic Space",
    labels={"x": "PC1", "y": "PC2", "z": "PC3", "color": "ESG Topic"},
    color_discrete_sequence=px.colors.qualitative.Set2
)
fig.update_traces(marker=dict(size=7, opacity=0.85))
fig.update_layout(height=650)
fig.show()
print(f" {len(vectors)} vectors plotted across {len(ESG_QUERIES)} topics")
```

```python
# QnA
def esg_qa(question, k=3):
    print(f"\n {question}")
    print("-" * 60)
    results = vectorstore.similarity_search(question, k=k)
    combined = " ".join([r.page_content for r in results])
    answer = summarize(combined, topic="the question asked")
    print(f" {answer}")
    print("\n📎 Sources:")
    for i, r in enumerate(results, 1):
        print(f"  [{i}] Page {r.metadata.get('page','?')}: {r.page_content[:100]}...")

questions = [
    "What are the company's Scope 1 and Scope 2 greenhouse gas emissions?",
    "What physical climate risks does the company face?",
    "What are the renewable energy targets and progress?",
    "How does the company manage supply chain sustainability risks?"
]

for q in questions:
    esg_qa(q)
```

```python
# Export to EXcel
rows = []
for q in questions:
    results = vectorstore.similarity_search(q, k=3)
    combined = " ".join([r.page_content for r in results])
    answer = summarize(combined)
    for i, r in enumerate(results, 1):
        rows.append({
            "Question": q,
            "AI Answer": answer,
            "Source #": i,
            "Page": r.metadata.get('page', '?'),
            "Preview": r.page_content[:120].strip()
        })

df = pd.DataFrame(rows)
df.to_excel("Tesla_esg_qa_output.xlsx", index=False)
print("Saved")
print(df[["Question", "AI Answer", "Page"]].to_string(index=False))
```
